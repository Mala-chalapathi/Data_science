```python
import pandas as pd

def summarize_data(df, column_names, funs, group_by):
    # Dictionary to map function names to pandas aggregation functions
    agg_funcs = {
        "sum": "sum",
        "mean": "mean",
        "max": "max",
        "min": "min",
    }

    # Validate that all functions in funs are supported
    unsupported_funcs = [func for func in funs if func not in agg_funcs]
    if unsupported_funcs:
        raise ValueError(f"Unsupported functions: {unsupported_funcs}. Supported functions are: {list(agg_funcs.keys())}")

    # Define the aggregation dictionary
    aggregation = {}
    for col in column_names:
        for func in funs:
            aggregation[f"{col}_{func}"] = (col, agg_funcs[func])

    # Group by and aggregate
    result = (
        df.groupby(group_by)
        .agg(**aggregation)
        .reset_index()
    )

    return result
```

```python
result=summarize_data(df,column_names=['read','write','math','science','socst'],funs=['sum','mean','max','min'],group_by='female')
print(result)
```
